## Histogram

A histogram represents the distribution of a variable by creating bins (interval) along the range of values and showing vertical bars to indicate the number of observations in each bin. 

For example, let's visualize the distribution of values of sepal width in the flowers dataset. We can use the `plt.hist` function to create a histogram.

### bins statictical term
### interval mathematical term

```python
import matplotlib.pyplot as plt 
import seaborn as sns

sns.set_style("darkgrid")

flower_ds = sns.load_dataset("iris")

flower_ds.petal_length
```

```python
flower_ds.describe()
```

```python
plt.title("Distribution of Sepal Width")

plt.hist(flower_ds.sepal_width);
```

```python
#specify the number of bins
plt.hist(flower_ds.sepal_width,bins=5)
```

### Increase the readability of by the help of numpy

```python
import numpy as np

np.arange(2,5,0.25)
```

```python
#Specify the bounderies of each bins

plt.hist(flower_ds.sepal_width,bins=np.arange(2,5,0.25));
```

```python
#Bin of unequal sizes
plt.hist(flower_ds.sepal_width,bins=[1,3,4,4.5])
```

### Multiple Histograms

Similar to line charts, we can draw multiple histograms in a single chart. We can reduce each histogram's opacity  so that one histogram's bars don't hide the others'.

Let's draw separate histograms for each species of flowers.

```python
setosa_df = flower_ds[flower_ds.species == 'setosa']
versicolor_df = flower_ds[flower_ds.species == 'versicolor']
virginica_df = flower_ds[flower_ds.species == 'virginica']
```

```python
plt.hist(setosa_df.sepal_width,alpha=0.4,bins=np.arange(2,5,0.25))
plt.hist(versicolor_df.sepal_width,alpha=0.5,bins=np.arange(2,5,0.25))
plt.legend(['Setosa','Versicolor']);
```

#### We can also stack multiple histograms on top of one another.

```python

plt.figure(figsize=(12,6))
plt.title('Distribution of Sepal Width')
plt.hist([setosa_df.sepal_width,versicolor_df.sepal_width,virginica_df.sepal_width]
         ,bins=np.arange(2,5,0.25),stacked=True)
plt.legend(["Setosa","Versicolor","Virginica"]);
```
