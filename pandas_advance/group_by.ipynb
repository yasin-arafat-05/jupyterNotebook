# 5. Apply: 

```python

# we already do this: df.groupby('Genre).min().head(2)
# apply(group)-> we get groups as an parameters
# But in apply we can send our custom function.
df.groupby('Genre').apply(min).head(2)
```

```python

#Q1: find the number of movies Start with 'A' in each Genre:
def foo(group_data):
    cnt = 0 
    for title in group_data["Series_Title"]:
        if title[0]=="A":
            cnt+=1 
    return cnt 

def shortfoo(group_data):
    return group_data["Series_Title"].str.startswith('A').sum()

df.groupby('Genre').apply(shortfoo)
    
```

```python
#
#Q2: add a new column normalize rating based on each group: 
#like action grup: normalize_rating = (X_rating) - X_min_rating / (X_max_rating + X_min_rating)
#
def normalize_rating(group_data):
    group_data["Norm_Rating"] = (group_data["IMDB_Rating"] - group_data["IMDB_Rating"].min()) /(group_data["IMDB_Rating"].max()-group_data["IMDB_Rating"].min()+)
    return group_data
df.groupby('Genre').apply(normalize_rating)
```

# 6. GroupBy on Multple Columns: 


```python

# Director: Akira and Star: Toshirô, They do total 7 movies together.
df.groupby(['Director','Star1']).size().sort_values(ascending=False)
```

```python

# Q1: get a particular group value: 
# now in get_group we need to pass tuple 
df.groupby(['Director','Star1']).get_group(('Christopher Nolan','Christian Bale'))
```

```python

#Q2: Find the most earning actor-director combo:
df.groupby(['Director','Star1'])['Gross'].max().sort_values(ascending=False).head(1)
```

```python

#Q3: Find the best(metascore(avg)) actor-genre combo:
df.groupby(['Director','Star1'])["Metascore"].mean().reset_index().sort_values(by='Metascore',ascending=False).head(1)
```

```python

#Q3: aggerate function on group by on multiple columns:
df.groupby(['Director','Star1']).agg({
    'Runtime':['min','max']
})
```
