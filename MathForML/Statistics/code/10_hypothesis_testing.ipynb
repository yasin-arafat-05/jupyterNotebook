### **Problem 1**

A factory claims that the average lifetime of its bulbs is **1200 hours**.
A random sample of **49 bulbs** shows a mean lifetime of **1160 hours**.
The population standard deviation is known to be **140 hours**.
Test the claim at **α = 0.05**.


````python

# Given,
# population mean = 1200
# sample size = 49 
# sample mean = 1160 
# population std = 140 
# alpha = 0.05 
# As Population std is given then we will apply Z-Score apporach.

# Null Hypothesis: population mean = 1200 
# Alternate Hypothesis : population mean not equal to 1200 (Two tailed test)
sample_mean = 1160 
sample_std = 140 
population_mean = 1200
sample_size = 49 
alpha = 0.05

z_score = (sample_mean-population_mean)/(np.sqrt(sample_std**2/sample_size))

# multiplied by 2 for two tailed test: 
p_value = 2* (1 - norm.cdf(z_score)) 

if p_value>alpha:
    print("Null Hypothesis is Accepted.")
    print("population mean is equal to 1200")
else: 
    print("Null Hypothesis is Rejected.")
    print("population mean not equal to 1200")

````


### **Problem 2**

Two machines produce the same type of component.

* Machine A: n = 15, mean = 52, standard deviation = 4
* Machine B: n = 12, mean = 48, standard deviation = 5

At the **5% level of significance**, determine whether there is a difference in average output.



````python

# Given, 
# n1 = 15 
# n2 = 12 
# std1 = 4 
# std2 = 5 
# alpha = 0.05 
# As, std is given we will apply Z-score
n1 = 15 
n2 = 12 
std1 = 4 
std2 = 5 
mean1 = 52 
mean2 = 48 
alpha = 0.05 

Null_hypo = "(Population Mean)_1 = (Population Mean)_2"
# two tailed test
Alter_hypo = "(Population Mean)_1 != (Population Mean)_2"

s_E_X = np.sqrt(((std1**2/n1)+ (std2**2/n2)))
z_score = (mean1-mean2)/s_E_X
p_value = 2*(1-norm.cdf(z_score))

print("z_score: ",z_score)
print("p_value: ",p_value)
if p_value<alpha:
    print("Null Hypothesis is Rejected: {}".format(Alter_hypo))
else: 
    print("Null Hypthesis is Accepted: {}".format(Null_hypo))
    
    
````


### **Problem 6**

Before and after installing a new software, the execution time (in seconds) of the same 8 programs is recorded:

| Before | After |
| ------ | ----- |
| 12     | 10    |
| 15     | 13    |
| 11     | 9     |
| 14     | 12    |
| 13     | 11    |
| 16     | 14    |
| 10     | 9     |
| 12     | 10    |

Test whether the new software improves performance at **α = 0.05**.


````python

# n<30: t-test and and not independent(paired-t-test)
data = {
    "Before": [12,15,11,14,13,16,10,12],
    "After":  [10,13,9,12,11,14,9,10]
}
df = pd.DataFrame(data,index=[1,2,3,4,5,6,7,8])
num_of_sample = 8 

# One-tail(Left-Tail) Test:
Null_hypo = "No Improvement, mean1 = mean2"

""" 
Software perform well means: Execution time will be less
"""
Alter_hypo = "Improves,mean2<mean1"
alpha = 0.05 

df["d=After-Before"] = df["After"]-df["Before"]
df["d^2"] = df["d=After-Before"]*df["d=After-Before"]
d_bar = df["d=After-Before"].mean()

# calcuate: modified from of std.
S = np.sqrt((df["d^2"].sum() - (d_bar**2*num_of_sample))/(num_of_sample-1))

# calculate degree of freedom: 
df = num_of_sample - 1 

# caclualte t and p value
t_val = ((d_bar)/S) * np.sqrt(num_of_sample)

# left-tailed-test
p_value = t.cdf(t_val,df)


print("t_value: ",t_val)
print("p_value: ",p_value)

if p_value<alpha:
    print("Null Hypothesis is Rejected. {}".format(Alter_hypo))
else: 
    print("Null Hypothesis is Accepted. {}".format(Null_hypo))
    
````

### **Problem 10**

The number of customers visiting a store on different days is recorded as:

| Day      | Mon | Tue | Wed | Thu | Fri |
| -------- | --- | --- | --- | --- | --- |
| Observed | 48  | 52  | 50  | 46  | 54  |

Test whether customer visits are **uniform across days** at **α = 0.05**.

````python
from scipy.stats import chi2

# chi-square goodness of fit: Observe and Expected value: 
data = {
    "Day":['Mon','Tue','Wed','Thu','Fri'],
    "Observed":[48,52,50,46,54],
}
df = pd.DataFrame(data,index=[i for i in range(1,6)])

# right-tailed test
Null_hypo = "Customer visits are uniform across days"
Alter_hypo = "Customer visits are not uniform across days"

alpha = 0.05

total = df["Observed"].sum()
k = len(df)
df["Expected"] = total/k 


df['(O-E)^2'] = (df["Observed"]-df["Expected"])**2 
df['(O-E)^2/E'] = df['(O-E)^2']/df["Expected"]
chi_square_value = df['(O-E)^2/E'].sum()

#find degree of freedom:
dof = len(df)-1 

p_value = 1 -  chi2.cdf(chi_square_value,dof)

if p_value < alpha:
    print("Reject H0 →", Alter_hypo)
else:
    print("Fail to reject H0 →", Null_hypo)
    
````
