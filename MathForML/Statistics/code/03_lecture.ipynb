```python
import random
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python

# number between (1 to 6)
random.randint(1,6)
```

```python

L = []
for i in range(10000):
    L.append(random.randint(1,6))
    
print(len(L)) # We have 10000 Different value:
print(L[:5])  # First, 10 value: 
```

```python
df = pd.Series(L)
df
```

```python
sns.countplot(df.value_counts())
```

# `#1.2 : Probability Mass Funtion: (Two dice:) `
# `#1.3 : Cumulative Distrubition Function: `


<Br>
<Br>
<Br>

---

<Br>

---

<Br>

---


```python
L = []

for i in range(1000):
    a = random.randint(1,6)
    b = random.randint(1,6)
    L.append(a+b)

print(len(L))
print(L[:5])
```

```python
s = (pd.Series(L).value_counts()/pd.Series(L).value_counts().sum()).sort_index()
s
```

```python
s.plot(kind="bar") # bar graph: 
```

```python
np.cumsum(s).plot(kind="bar") # cumualtive graph: (CDF)
```
