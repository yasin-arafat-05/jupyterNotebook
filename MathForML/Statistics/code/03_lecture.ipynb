```python
L = []

for i in range(1000):
    a = random.randint(1,6)
    b = random.randint(1,6)
    L.append(a+b)

print(len(L))
print(L[:5])
```

```python
s = (pd.Series(L).value_counts()/pd.Series(L).value_counts().sum()).sort_index()
s
```

```python
s.plot(kind="bar") # bar graph: 
```

```python
np.cumsum(s).plot(kind="bar") # cumualtive graph: (CDF)
```

# `#1.4 : Parametric Density Estimator: `

<Br>
<Br>
<Br>

---

<Br>

---

<Br>

---


```python
"""
We, generate some random (normal data(population) of Normal Distribution) 
Whose,
Mean near around = 50
Standard Deviation = 5
"""
sample = np.random.normal(loc=50,scale=5,size=1000)

print()
print("sample mean: ",sample.mean())
print("same standard deviation: ",sample.std())
print()
```

```python
plt.hist(x=sample,bins=10,color="green") # my data is normal:
```

<br>
<br>

## `# see, the parameter of: normal Distrubition: `

[PDF_Parametric_Density_Estimator_SEE_THE_PDF_FORMULA](https://en.wikipedia.org/wiki/Normal_distribution)

<br>
<br>

```python

sample_mean = sample.mean()
sample_std = sample.std()
```

```python

from scipy.stats import norm
dist = norm(sample_mean,sample_std)
```

```python

values = np.linspace(sample.min(),sample.max(),100)
print(values)
print("len of values: ", len(values))
```

```python
probabilitites = [dist.pdf(value) for value in values]
len(probabilitites)
```

```python

#plot the histrogram and pdf
plt.hist(sample,bins=10,density=True) # draw the distrogram 
plt.plot(values,probabilitites) # draw the PDF(from weikipidea websites)
```

<br>
<br>

## `#Change the mean: `

If we change the mean then our pdf will be change: 

<br>
<br>

```python

from scipy.stats import norm
print("before: sample mean: ",sample_mean)
print("before: sample mean: ",sample_std)
sample_mean = 30
sample_std = 10
dist = norm(sample_mean,sample_std)
probabilitites = [dist.pdf(value) for value in values]
values = np.linspace(sample.min(),sample.max(),100)

plt.hist(x=sample,bins=10,density=True)
plt.plot(values,probabilitites)
```
