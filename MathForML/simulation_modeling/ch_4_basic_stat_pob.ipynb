```python

# =========== Book Question: 4.4 ===========
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def p(x,y):
    if x in [0,1,2] and y in [0,1,2,3]:
        return (x+y)/30
    return 0

x_vals = [0,1,2]
y_vals = [0,1,2,3]

joint = np.zeros((len(x_vals), len(y_vals))) 

for i,x_val in enumerate(x_vals):
    for j,y_val in enumerate(y_vals):
        joint[i,j] = p(x_val,y_val)

df = pd.DataFrame(joint)

#1.Heatmaps
print(joint)
sns.heatmap(data=df,annot=True)


#2. Marginal
p_x = df.sum(axis=1)
p_y = df.sum(axis=0)
print(f"\nMarginal Probability Px(X):\n",p_x)
print(f"\nMarginal Probability Py(X):\n",p_y)

#3. CDF:
f_x = np.cumsum(p_x)
f_y = np.cumsum(p_y)
print("\nCDF of F_X:\n",f_x)
print("\nCDF of F_Y:\n",f_y)


# plot CDF:
fig,ax = plt.subplots(1,2,figsize=(10,3))


ax[0].step(x=x_vals,y=f_x,where="mid")
ax[0].plot(x_vals,f_x,'o')
ax[0].set_title("CDF of F_X")
ax[0].set_xlabel("X-value(x)")
ax[0].set_ylabel("F_X(x)")
ax[0].grid(True)
ax[0].set_xlim(-2,5)
ax[0].set_ylim(0,1.1)


ax[1].step(x=y_vals,y=f_y,where="mid")
ax[1].plot(y_vals,f_y,'o')
ax[1].set_title("CDF of F_Y")
ax[1].set_xlabel("X-value(x)")
ax[1].set_ylabel("F_Y(x)")
ax[1].grid(True)
ax[1].set_xlim(-2,5)
ax[1].set_ylim(0,1.1)



plt.show()


```

```python

# =========== Book Question: 4.7 ===========
import numpy as np
import matplotlib.pyplot as plt

# f_X(x)
x = np.linspace(0, 1, 100)
f_x = 1.5 - x
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(x, f_x, color='blue')
plt.title('Marginal PDF f_X(x)')
plt.xlabel('x')
plt.ylabel('f_X(x)')
plt.grid(True)

# f_Y(y)
y = np.linspace(1, 2, 100)
f_y = y - 0.5
plt.subplot(1, 2, 2)
plt.plot(y, f_y, color='green')
plt.title('Marginal PDF f_Y(y)')
plt.xlabel('y')
plt.ylabel('f_Y(y)')
plt.grid(True)

plt.tight_layout()



# F_X(x)
x_full = np.linspace(-0.5, 1.5, 200)
f_x_cum = np.where(x_full < 0, 0, np.where(x_full >= 1, 1, 1.5*x_full - 0.5*x_full**2))
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(x_full, f_x_cum, color='blue')
plt.title('CDF F_X(x)')
plt.xlabel('x')
plt.ylabel('F_X(x)')
plt.grid(True)

# F_Y(y)
y_full = np.linspace(0.5, 2.5, 200)
f_y_cum = np.where(y_full < 1, 0, np.where(y_full >= 2, 1, 0.5*y_full**2 - 0.5*y_full))
plt.subplot(1, 2, 2)
plt.plot(y_full, f_y_cum, color='green')
plt.title('CDF F_Y(y)')
plt.xlabel('y')
plt.ylabel('F_Y(y)')
plt.grid(True)

plt.tight_layout()
plt.show()
```
