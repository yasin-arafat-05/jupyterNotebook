```python

# Now we will extract feature  from the data to remove the missing data: 


X_train = train.drop(columns="SalePrice")  # in the matrix form

y_train = train["SalePrice"]  #in the vector form  that's why we keep the y small (different from X_train)


# Print the shape of train and test
print(f"X_train data's shape: {X_train.shape}")
print(f"y_train data's shape: {y_train.shape}")
```


# Numerical Missing Value Imputation

```python

# store the numerical value in num_vars
num_vars = X_train.select_dtypes(include=["int64","float64"]).keys()
num_vars
```

```python

# count the missing value number 

X_train[num_vars].isnull().sum()
```

<br>

এখন, মনে করি যখন কোন user আমাদের কে ডাটা দিবে তখন যদি যেই ডাটা তে LotArea এখন তো শূন্য হয়? তখন আমরা কি করবো !
আমাদের যে train dataset থাকবে সেইগুলোকে আমরা সেইখানে বসিয়ে দিব । 
এর জন্য আমরা sklearn  ব্যবহার করবো ।  
<br>

```python
imputer_mean = SimpleImputer(strategy="mean")

# এখন imputer_mean  object এ যেই ডাটা বা (feature) সেট দিব সেইটার (সেই feature গুলোর ) mean ভ্যালু পাবো । 

# .fit() method সকল numerical value গুলোর mean বের করে নিজের কাছে রেখে দিবে আমাদের দরকার হলে আমরা সেইগুলো 
# ব্যবহার করতে পারবো । বা object create করে নিজের কাছে রেখে দিবে । 

imputer_mean.fit(X_train[num_vars]) 


# আমি numerical যত গুলো row দিয়েছি(num_vars) এর মধ্যে সব গুলোর mean ভ্যালু return করবে । *staticstics_ property*

imputer_mean.statistics_

# কিন্তু real ডাটা এ সেইখানে value missing  আছে সেইখানে  আমরা এই গুলো ব্যবহার করিনি । তাই আমাদের next কাজ হচ্ছে 
# ভ্যালু গুলো কে সেই খানে add করা । 
# এখন আমরা transform method করে ভ্যালু গুলো কে missing value তে impute করবো । 
```

```python

# sklearn এর সবচেয়ে বড় disadvantage হচ্ছে আমরা দিলাম pandas dataframe কিন্তু পেলাম একটা array.

imputer_mean.transform(X_train[num_vars])
```

```python
#  X_train[num_vars] এর মধ্যে থাকা data গুলো কে imputer_mean.transform(X_train[num_vars]) এর ভ্যালু গুলো দিয়ে update করবো ।  

X_train[num_vars] = imputer_mean.transform(X_train[num_vars])

# ultimetly we update in test data set.

test[num_vars] = imputer_mean.transform(test[num_vars])
```

```python

# Now check in X_train is there is still have nan value 
X_train[num_vars].isnull().sum()
```

```python
#   Similarly    for     test   data

test[num_vars].isnull().sum()
```

<br> <br>

উপরের মতো আমাদের numerical value গুলো নিজের ইচ্ছে মতো mean,median,mode করে update করতে পারবো । 
কিন্তু, আমাদের যদি এখন, নিজের ইচ্ছে মতো value দিতে হয় তাই হলে আমরা,
### imputer_mean = SimpleImputer(strategy='constant',fill_value=99) 
fill_value নিজের ইচ্ছে  মতো দিতে পারি।

# `Missing value imputation for categorial value:`

<br> <br>

```python
cat_var = X_train.select_dtypes(include='object').columns
cat_var
```

```python
# Count the missing values:

X_train[cat_var].isnull().sum()
```

```python

imputer_mode = SimpleImputer(strategy='most_frequent')
```

```python
 # give value in imputer_mode object
imputer_mode.fit(X_train[cat_var])

# see the most frequent value in an array
imputer_mode.statistics_
```

```python
# transform the data set 

X_train[cat_var] = imputer_mode.transform(X_train[cat_var])

test[cat_var]  = imputer_mode.transform(test[cat_var])
```

```python
# Check the null value: 
X_train[cat_var].isnull().sum()
```

```python
test[cat_var].isnull().sum()
```
