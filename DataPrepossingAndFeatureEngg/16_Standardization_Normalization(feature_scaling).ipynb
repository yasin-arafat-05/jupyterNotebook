# Why standardization?

`Standarization is a part of feature caling.`

![Alt text](image-7.png)

- আমরা এইটা করে থাকি যাতে mean = 0 and standard_deviation = 1 হয় । 

- যদি orginal dataset এর distribution normal হয় তাহলে এর standardization এর distribution ও normal হবে । আর যদি orginal dataset এর distribution skewed হয় তাহলে এর standardization এর distribution ও skewed হবে । 

![Alt text](image-8.png)

যখন  dataset এর curve উপরের মতো হবে তখন আমরা বলবো আমাদের dataset  normal distribution বা gaussian distribution ফলো করে ।

`statistic এ আমরা এইগুলো আরো ভালোভাবে দেখে এসেছি ।`

# Why Normalization?

![Alt text](image-9.png)

# Code Example:

---

<br> <br>



```python
import pandas as pd 
import numpy as np 
import seaborn as sns

from sklearn.preprocessing import StandardScaler, MinMaxScaler

from sklearn.model_selection import train_test_split
```

```python

# we use seaborn to use the titanic dataset

df = sns.load_dataset("titanic")
df.shape
```

```python
df.info()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# আমরা সব গুলো feature নিয়ে কাজ করবো না । 

df2 = df[["survived","pclass","age","parch"]]

# এখানে, survived আমাদের target variable বা dependent variable.

df2
```

```python

# check nan value in my new dataset
df2.isnull().sum()
```

```python

# fill the nan value with mean(as dtype is int64):
df3 = df2.fillna(df2.mean())
df3.isnull().sum()
```

```python

X = df3.drop("survived",axis=1)
y = df3["survived"]

print(f"Shape of X: {X.shape}")
print(f"Shape of y: {y.shape}")
```

```python

# next video তে train test split নিয়ে পড়বো ।

# test_size আমাদের test data 20% হবে main dataset এর । 
X_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=51)

print(f"Shape of X_train: {X_train.shape}")
print(f"Shape of X_test: {x_test.shape}")

print(f"Shape of y_train: {y_train.shape}")
print(f"Shape of y_test: {y_test.shape}")


```


---

# এখন আমাদের ডাটা train, test এর জন্য প্রস্তুত । এখন আমরা feature scaling ব্যবহার করতে পারি । 

---


```python

sc = StandardScaler()

sc.fit(X_train)
```

```python
# mean value
sc.mean_
```

```python
# standard daviation
sc.scale_
```

```python
# we can check it in: 
X_train.describe()
```


--- 

# আর আমরা feature scaling train data এর উপর apply করি । আর train and test data এর উপর আমরা যেই feature scaling করলাম সেইটা apply করবো বা feature transform করবো।

--- 

#### এখন আমাদের feature scaling শেষ StandardScaler ব্যবহার করে । এখন আমাদের feature transform করতে হবে । 

```python
X_train = sc.transform(X_train)
X_test = sc.transform(x_test)
```
