```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
pd.set_option('display.max_rows',None)
df = sns.load_dataset("tips")
```

```python
#info of male an female
female = df[df["sex"]=="Female"]
male = df[df['sex']=='Male']
```

```python
print(female.info())

print()
print()

print(male.info())

print()
print()

print(female.columns)

print()
print()

print(male.columns)
```

```python
plt.figure(figsize=(20,20))

plt.subplot(2,2,1)
plt.title("Histplot of Females")
sns.histplot(female)

plt.subplot(2,2,2)
plt.title("Histplot of Males")
sns.histplot(male)

plt.subplot(2,2,3)
plt.title("Total bill of Female")
sns.histplot(female['total_bill'],color='m')

plt.subplot(2,2,4)
plt.title("Total bills  of Males")
sns.histplot(male['total_bill'],color='y');
```
