# To Remove Data We Need Know The Parcentage Of Missing Data

```python
per = df.isnull().sum()/df.shape[0]

per = per*100
per
```

#### In Our Data Set Of Percentage (We delete the column whose percentage of data is greater than 20)

```python
#In this case we are conserding >=15

drop_columns = per[per>=15].keys()

drop_columns
```

```python
print(f"Before droping: {df.shape}")
df2 = df.drop(columns=drop_columns,axis=1)
print(f"After droping: {df2.shape}")
```

```python
plt.title("-    Before:     -")
sns.heatmap(df.isnull());
```

```python
#Now Observe The HeatMap Plot
plt.title("-    After    -")
sns.heatmap(df2.isnull());
```

# Drop Rows: 

```python
#remove the rows which contains NaN vlaues

df3 = df2.dropna()

print(f"In Df: {df.shape}")
print(f"In Df2: {df2.shape}")
print(f"In Df3: {df3.shape}")
```

```python
df3_histplot = df3.select_dtypes(include=['int64','float64']).columns
df3_histplot
```

```python
#show the histogram plot before and after data clearning.

sns.histplot(df['MSSubClass'],alpha=0.9,color="yellow")
sns.histplot(df3['MSSubClass'],alpha=0.9,color="green");
```

##### Now, We Need To Compare all the histogram we done before. 

```python
# By the help of subplot and enumerate

plt.figure(figsize=(25,25))

plt.subplots_adjust(
    left=0.1,
    bottom=0.1, 
    right=0.9, 
    top=0.9, 
    wspace=0.4, 
    hspace=0.4
)

for i,var in enumerate(df3_histplot):
    plt.subplot(9,4,i+1)
    sns.histplot(df[var],alpha=0.9,color="yellow")
    sns.histplot(df3[var],alpha=0.9,color="green");
    
```
