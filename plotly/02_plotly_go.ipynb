# `#01: Scatter Plot`

<br>

- **Problem:** We are going to draw a scatter plot between Batsman Avg(X axis) and Batsman Strike Rate(Y axis) of the top 50 batsman in IPL(All time).

<br>

```python

ipl.columns
```

```python
len(ipl["batter"])
```

```python
len(ipl["batter"].unique())
```

```python

top50 = ipl.groupby(by='batter')['batsman_runs'].sum().sort_values(ascending=False).head(50).index.tolist()
top50
```

```python

# top 50 batsman
new_ipl = ipl[ipl['batter'].isin(top50)]
new_ipl
```

```python
new_ipl.columns
```

```python



# calculating strike rate: 
# (no. of run score) / (no. of ball played)
runs = new_ipl.groupby('batter')['batsman_runs'].sum()
balls = new_ipl.groupby('batter')['batsman_runs'].count()
sr = (runs/balls)*100
sr = sr.reset_index() # convert into dataframe
sr

```

```python

# avareage: 
# AVG = (total_number_of_runs) / (No of outs)
out = ipl[ipl["player_dismissed"].isin(top50)]
nouts = out['player_dismissed'].value_counts()
avg = runs/nouts 
avg = avg.reset_index()
avg.columns = ['batter','avg']
sr = sr.merge(avg,on='batter')
sr
```

```python

sr["batter_code"] = sr["batter"].astype('category').cat.codes


# now finally draw the scatter pot:
trace = go.Scatter(x=sr["avg"],y=sr["batsman_runs"],mode='markers',
                   marker={'color':"#020a18",'size':12})
data = [trace]
layout = go.Layout(
    title="AVG vs RUNS",
    xaxis=dict(title="Batter AVG"),
    yaxis=dict(title="Batter Runs"),
    
)
fig = go.Figure(data,layout)
fig.show()
```

```python


fig = px.scatter(
    data_frame=sr,
    x="avg",
    y="batsman_runs",
    title="avs vs runs",
    color="batter", 
)


fig.update_traces(
    marker=dict(
        size=12,
        symbol='circle',
        opacity=0.9
    )
)

fig.show()

```

# `#02: Line Plot`

<br>

**Select the player Virat find his performance into a line chart**

<br>

```python


virat_performance = ipl[ipl['batter']=='V Kohli'].groupby('season')['batsman_runs'].sum().reset_index()
virat_performance
```

```python



# mode = lines+markers:)
trace = go.Scatter(x=virat_performance['season'],
                   y=virat_performance['batsman_runs'],
                   mode="lines+markers",
                   marker={"size":12})
data = [trace]
layout = go.Layout(
    title="Virat Kholi Performance Over The Years",
    xaxis=dict(title="Years"),
    yaxis=dict(title="run of virat kholi")
)
fig = go.Figure(data,layout)
fig.show()


```

```python
dir(go.Layout)
```

```python


fig = px.line(data_frame=virat_performance, x="season", y="batsman_runs", template='plotly_dark')
fig.update_traces(
    mode = "lines+markers",
    marker=dict(size=12, color='#00cc96')
)

fig.update_layout(
    xaxis=dict(title='Year', showgrid=True),
    yaxis=dict(title="Virat Runs During that Year", showgrid=True),
    font=dict(size=12, color='white'),
    
    # figure oclor
    paper_bgcolor='#0e1117',
    plot_bgcolor='#0e1117',
)

fig.show()
```
